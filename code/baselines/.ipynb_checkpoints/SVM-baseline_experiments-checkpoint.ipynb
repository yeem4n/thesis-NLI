# SVM Baselines
The following code was used to implement the SVM baselines. For the VESPA and TOEFL11 dataset, we evaluate using a train/test split. For the ICLE-NLI dataset, we evaluate under stratified 5-fold cross-validation. We implement an SVM with BoW features, as well as an SVM with 1-9 N-grams with TF-IDF representation, and perform under-sampling for the VESPA dataset to deal with the imbalanced training set. 

```python
# import libraries
from sklearn.svm import LinearSVC
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from sklearn.model_selection import cross_val_predict, cross_val_score, StratifiedKFold
from sklearn.pipeline import make_pipeline
import numpy as np
from imblearn.under_sampling import RandomUnderSampler
```

```python
# define file paths--change accordingly
toefl_train = "./data/TOEFL11_2013/TOEFL-preprocessed/train.csv"
toefl_test = "./toefl_results.csv"
vespa_train = './data/VESPA/VESPA-train.csv'
vespa_test = './data/VESPA/VESPA-test.csv'
icle = "./ICLE-NLI-results-final.csv"
```

## VESPA

```python
# Load dataset
train_df = pd.read_csv(vespa_train)
tr_texts = train_df['text'].tolist()
tr_labels = train_df['language'].tolist()

test_df = pd.read_csv(vespa_test)
print(f'Number of samples: {len(test_df)}')
test_texts = test_df['text'].tolist()
test_labels = test_df['language'].tolist()
```

### SVM BoW

```python
model = make_pipeline(CountVectorizer(), LinearSVC()) # initiate Linear SVC with BoW count features
model.fit(tr_texts, tr_labels) 
svm_predictions = model.predict(test_texts) 
all_labels = sorted(list(set(test_labels))) # get all labels for the confusion matrix
cm = confusion_matrix(test_labels, svm_predictions) 
cm_display = ConfusionMatrixDisplay(cm, display_labels=all_labels).plot(cmap='Blues') 
acc = "{:.2f}".format(accuracy_score(test_labels, svm_predictions)*100)
print(acc)
```

## Under-sampled training set

```python
# perform random under-sampling
rus = RandomUnderSampler(sampling_strategy="not minority", random_state=0)
X = train_df.drop('language',axis=1)
y = train_df['language'].tolist()
X_resampled, y_resampled = rus.fit_resample(X, y)
# print(len(y_resampled))
# print(X_resampled['text'][0])
sample_tr_texts = X_resampled['text'].tolist()
sample_tr_labels = y_resampled

# train SVM with BoW count features
model = make_pipeline(CountVectorizer(), LinearSVC())
model.fit(sample_tr_texts, sample_tr_labels) 
svm_predictions = model.predict(test_texts) # test
all_labels_TOEFL = sorted(list(set(test_labels)))
cm = confusion_matrix(test_labels, svm_predictions)
cm_display = ConfusionMatrixDisplay(cm, display_labels=all_labels_TOEFL).plot(cmap='Blues')
acc = "{:.2f}".format(accuracy_score(test_labels, svm_predictions)*100)
print(acc)
```

## TOEFL11

```python
# Load dataset
train_df = pd.read_csv(toefl_train)
tr_texts = train_df['text'].tolist()
tr_labels = train_df['language'].tolist()

test_df = pd.read_csv(toefl_test)
print(f'Number of samples: {len(test_df)}')
test_texts = test_df['text'].tolist()
test_labels = test_df['language'].tolist()
```

### SVM BoW

```python
vectorizer=CountVectorizer() # bow vectorizer
tr_sents_vector_data = vectorizer.fit_transform(tr_texts) # transform training texts into bow

lin_classifier = LinearSVC()
lin_classifier.fit(tr_sents_vector_data, tr_labels) 

test_vector_data = vectorizer.transform(test_texts).toarray() # transform test into bow
svm_predictions = lin_classifier.predict(test_vector_data) 

cm = confusion_matrix(test_labels, svm_predictions)
all_labels_TOEFL = sorted(list(set(test_labels)))
cm_display = ConfusionMatrixDisplay(cm, display_labels=all_labels_TOEFL).plot(cmap='Blues')
acc = "{:.2f}".format(accuracy_score(test_labels, svm_predictions)*100)
print(acc)
```

### SVM 1-9 TF-IDF

```python
model = make_pipeline(TfidfVectorizer(analyzer='char', ngram_range=(1, 9)), LinearSVC())
model.fit(tr_texts, tr_labels) 
svm_predictions = model.predict(test_texts)

cm = confusion_matrix(test_labels, svm_predictions)
all_labels_TOEFL = sorted(list(set(test_labels)))
cm_display = ConfusionMatrixDisplay(cm, display_labels=all_labels_TOEFL).plot(cmap='Blues')
acc = "{:.2f}".format(accuracy_score(test_labels, svm_predictions)*100)
print(acc)
```

## SVM for ICLE

```python
# Load dataset
icle = "./ICLE-NLI-results.csv"
df = pd.read_csv(icle)
icle_texts = df['text']
icle_gold_labels = df['language']
```

### SVM BoW

```python
model = make_pipeline(CountVectorizer(), LinearSVC())
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
scores = cross_val_score(model, icle_texts, icle_gold_labels, cv=skf, scoring='accuracy')
print(sum(scores)/len(scores)*100)
```

### SVM 1-9 gram TF-IDF

```python
# create a pipeline with TfidfVectorizer and LinearSVC
model = make_pipeline(TfidfVectorizer(analyzer='char', ngram_range=(1, 9)), LinearSVC())

# Use StratifiedKFold for multiclass classification
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

# Perform five-fold cross-validation
scores = cross_val_score(model, icle_texts, icle_gold_labels, cv=cv, scoring='accuracy')
print(sum(scores)/len(scores)*100)
```
